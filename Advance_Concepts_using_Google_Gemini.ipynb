```python
from google.colab import userdata
import google.generativeai as genai
from IPython.display import display
from IPython.display import Markdown
import pathlib
import textwrap
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))


# Example usage:
input_text = "This is a • sample text with bullet points."
result = to_markdown(input_text)

display(result)
```

#Generation Configuration

```python
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
model.generate_content("Tell me about the avengers").text
```

```python
to_markdown(model.generate_content("Tell me about the avengers").text)
```

```python
response = model.generate_content(
    "Tell me about the avengers",
    generation_config = genai.types.GenerationConfig(
        candidate_count = 1,
        stop_sequences=["xi"],
        max_output_tokens=20,
        temperature = 1.0
    )
)
```

```python
response
```

```python
response = chat.send_message("How about a more detailed explaination to a techie person",stream=True)
```

```python
for chunk in response:
  print(chunk.text)
  print("_"*80)
```

```python
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```

#Use Embeddings

```python
result=genai.embed_content(
    model = "models/embedding-001",
    content = "What is the meaning of life?",
    task_type = "retrieval_document",
    title = "Embedding of a single string"
)
```

```python
result["embedding"]
```

```python
len(result["embedding"])
```

```python
results=genai.embed_content(
    model = "models/embedding-001",
    content = [
        'What is the meaning of life?',
        'How much wood would a woodchuck chuck?',
        'How does the brain work?'
    ],
    task_type = "retrieval_document",
    title = "Embedding of a single string"
)
```

```python
for i in results["embedding"]:
  print(i)
  print(len(i))
```
